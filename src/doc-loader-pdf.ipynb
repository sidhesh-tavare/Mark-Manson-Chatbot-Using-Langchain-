```python
from langchain.document_loaders import TextLoader

loader = TextLoader(
    r"C:\Users\sidhe\Desktop\Machine Learning\Projects\LangChain Projects\Mark Manson\data\raw\Subtle-Art.txt",
    encoding="utf-8"  # Match encoding with file creation
)
data = loader.load()
print(data[0].page_content)

```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=5000,      # Set chunk size
    chunk_overlap=50,     # Set overlap
    length_function=len,
    is_separator_regex=False,  # No custom separator logic
)

# Step 3: Iterate over all pages and split them into chunks
all_chunks = []
print(len(data))
for doc in data:
    # Split the content of each document/page into chunks
    chunks = text_splitter.split_text(doc.page_content)
    all_chunks.extend(chunks)  # Add the chunks to the all_chunks list

```
